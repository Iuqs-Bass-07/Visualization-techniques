```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('iris')
```

```python
df.sample(5)
```

```python
df.describe()
```

```python
sns.set_style('darkgrid')
plt.title('sepal_width distribution')
plt.hist(df.sepal_width);
```

```python
sns.set_style('whitegrid')
plt.title('sepal_length distribution')
plt.hist(df.sepal_length);
```

# Contolling bin and size

```python
# number of bins can be set
plt.hist(df.sepal_width, bins=5);
```

```python
# this code will create a list of number strating from 2 to 5 with a space of 0.25
np.arange(2, 5, 0.25)
```

```python
plt.hist(df.sepal_width, bins= np.arange(2, 5, 0.25));
```

```python
# bins of different size can be set
plt.hist(df.sepal_width, bins=[1, 3, 4, 4.5]);
```

# Multiple Histograms
different histograms can be drawn on one another with different opacity

```python
setosa_df = df[df.species == 'setosa']
virginica_df = df[df.species == 'virginica']
versicolor_df = df[df.species == 'versicolor']
```

```python
plt.hist(setosa_df.sepal_width, bins = np.arange(2, 5, 0.25), alpha=0.4)
plt.hist(versicolor_df.sepal_width, bins = np.arange(2, 5, 0.25), alpha=0.4)

plt.legend(['setosa', 'versicolor']);
```

# Stacked Histogram
Many histograms can be stacks together if passes ad as list

```python
plt.title('sepal width distribution')

plt.hist([setosa_df.sepal_width, virginica_df.sepal_width, versicolor_df.sepal_width],
         bins = np.arange(2, 5, 0.25),
         stacked=True)

plt.legend(['setosa', 'virginica' ,'versicolor']);
```
