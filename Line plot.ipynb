```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Line Chart

```python
yield_apples = [0.895, 0.91, 0.919, .926, .929, 0.931]
```

```python
plt.plot(yield_apples);
```

Customize the X-axis

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, yield_apples);
plt.xlabel('year')
plt.ylabel('tons per hectare');
```

# Plot multiple lines

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('year')
plt.ylabel('tons per hectare');
```

# Chart Title and legend

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('years')
plt.ylabel('tons per hectare')

plt.legend(['apples', 'oranges']);
```

#Line Markers

```python
plt.plot(years, apples, marker='o', c='b', alpha=.5)
plt.plot(years, oranges, marker='x', ls='--' , c='y', )

plt.xlabel('years')
plt.ylabel('tons per hectare')
plt.title('crops production')

plt.legend(['apples', 'oranges']);
```

# FMT Argument in plt.plot  
Third argument in plt.plot can be set as fmt = '[marker][line][color]' and if line in fmt in not specified then plt will be of points only



```python
# 'o-r' --->   o is round marker , - is solid line , b is blue color
plt.plot(years, apples, 'o-b')
# 'x--r' --->   x is cross marker , -- is dashed line , r is red color
plt.plot(years, oranges, 'x--r' )

plt.xlabel('years')
plt.ylabel('tons per hectare')
plt.title('crops production')

plt.legend(['apples', 'oranges']);
```

```python
# 'or' --->   o is round marker , line is not specified then it will plot only points , b is blue color
plt.plot(years, apples, 'ob')
# 'x--r' --->   x is cross marker , -- is dashed line , r is red color
plt.plot(years, oranges, 'x--r' )

plt.xlabel('years')
plt.ylabel('tons per hectare')
plt.title('crops production')

plt.legend(['apples', 'oranges']);
```

# Changing the figure size

```python
plt.figure(figsize=(5, 5))

plt.plot(years, apples, 'o-r');
```
