```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_flowers = sns.load_dataset('iris')
```

```python
df_flowers.sample(5)
```

```python
df_flowers['species'].unique()
```

```python
plt.plot(df_flowers['sepal_length'], df_flowers['sepal_width']);
```

```python
sns.scatterplot(df_flowers['sepal_length'], df_flowers['sepal_width']);
```

# Adding hue

```python
sns.scatterplot(df_flowers['sepal_length'], df_flowers['sepal_width'], hue = df_flowers['species'], s=100);
```

```python
plt.figure(figsize=(5,5))
plt.title('flower dimesion')

sns.scatterplot(df_flowers['sepal_length'],
                df_flowers['sepal_width'],
                hue = df_flowers['species'],
                s=100);
```

# Plotting using Pandas Dataframe
seaborn support the dataframe of the pandas instead of passing the argument specifically one can write the column name as argument and in the end pass dataframe as data

```python
plt.title('flower dimesion')

sns.scatterplot('sepal_length',
                'sepal_width',
                hue = 'species',
                data = df_flowers);
```
